```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

# Data Cleaning

```python
#checking for missing values
data.isnull().sum()
```

```python
#filling the missing values in Age column
data['Age'].fillna(data['Age'].median(), inplace=True)
```

```python
#filling the missing values in cabin column
data['Cabin']=data['Cabin'].fillna('Unknown')
```

```python
#imputing the missing values in Embarked column
data['Embarked'].fillna(data['Embarked'].mode()[0],inplace=True)
```

```python
data.isnull().sum()
```

```python
#remove duplicates
data.drop_duplicates(inplace=True)
```

```python
# Convert categorical variables to appropriate data types
data.dtypes
```

```python
categorical_columns = ['Sex', 'Embarked', 'Pclass']
for column in categorical_columns:
    data[column] = data[column].astype('category')
```

# Feature Engineering

```python
# Extract titles from names
data['Title'] = data['Name'].str.extract(' ([A-Za-z]+)\.')
data.head()
```

```python
# Create fare ranges
data['FareRange'] = pd.qcut(data['Fare'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])
data.head()
```

```python
# Create FamilySize feature
data['FamilySize'] = data['SibSp'] + data['Parch'] + 1
data.head()
```

```python
# Create the count plot
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='Title', hue='Survived')

# Rotate the x-axis labels
plt.xticks(rotation=45, ha='right')

# Set the title
plt.title('Survival by Title')

# Show the plot
plt.show()
```

```python
sns.countplot(data=data, x='FareRange', hue='Survived')
plt.title('Survival by Fare Range')
plt.show()
```

```python
sns.barplot(data=data, x='FamilySize', y='Survived')
plt.title('Survival by Family Size')
plt.show()
```

```python
# Calculate key statistics
survival_rate = data['Survived'].mean() * 100
female_survival_rate = data[data['Sex'] == 'female']['Survived'].mean() * 100
male_survival_rate = data[data['Sex'] == 'male']['Survived'].mean() * 100
first_class_survival_rate = data[data['Pclass'] == 1]['Survived'].mean() * 100
third_class_survival_rate = data[data['Pclass'] == 3]['Survived'].mean() * 100
average_age = data['Age'].mean()
average_fare = data['Fare'].mean()


print("\nKey Insights:")
print(f"Overall Survival Rate: {survival_rate:.2f}%")
print(f"Female Survival Rate: {female_survival_rate:.2f}%")
print(f"Male Survival Rate: {male_survival_rate:.2f}%")
print(f"First Class Survival Rate: {first_class_survival_rate:.2f}%")
print(f"Third Class Survival Rate: {third_class_survival_rate:.2f}%")
print(f"Average Age: {average_age:.2f}")
print(f"Average Fare: {average_fare:.2f}")
```

# Insights and Conclusions

1-SURVIVAL RATES:

a-Approximately 38% of passengers survived the disaster. b-Females had a significantly higher survival rate compared to males (~74% of females survived, while only ~19% of males survived).

2-PASSENGER CLASS (Pclass):

a-Passengers in 1st class had the highest survival rate (~63%), followed by 2nd class (~47%), and 3rd class (~24%). b-The likelihood of survival decreases with a lower passenger class.

3-AGE DISTRIBUTION

a-Most passengers were young adults between 20-40 years. b-Children under 10 had a relatively higher survival rate, indicating priority given to children during evacuation.

4-GENDER AND CLASS INTERATION:

a-Females in 1st and 2nd class had the highest survival rates (~97% and ~92%, respectively). b-Males in 3rd class had the lowest survival rate (~14%).

5-FARE

a-Higher fares were strongly correlated with higher survival rates. Passengers who paid higher fares (1st class) had better access to lifeboats and safety.

6-FAMILY SIZE

a-Passengers with a family size of 2-4 (e.g., small families) had higher survival rates. Those traveling alone (family size = 1) or in large families (family size > 4) had lower survival rates.
